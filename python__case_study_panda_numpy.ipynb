# Pandas Problems


```python
import pandas as pd
```

##1. Basic Exploration

```python
df = pd.read_csv("/auto-mpg.csv")
```

```python
df.head(10)   #Display first 10 rows
```

```python
df.shape  #rows & columns
```

```python
df.describe()   #Summary statistics
```

##2. Filtering and Indexing

```python
df[(df['weight'] < 3000) & (df['model year'] > 75)][['car name', 'mpg', 'weight']]
```

##3. Handling Missing Data

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)
```

```python
df.info()
```

##4. Data Transformation

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
```

```python
df[['horsepower', 'weight', 'power_to_weight_ratio']].head()
```

##5. Group By


```python
df.groupby('origin', as_index=False)['mpg'].mean()
```

##6. Sorting


```python
top10_cars=df.sort_values(by='mpg', ascending=False).head(10)
print(top10_cars)
```

```python
print(top10_cars['car name'])
```

##7. Apply function

```python
def performance_score(row):
  return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
print(df[['mpg', 'acceleration', 'weight', 'performance_score']].head())
```

##8. Visualization Preparation

```python
average = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
average.rename(columns={
    'mpg': 'avg_mpg',
    'weight': 'avg_weight',
    'horsepower': 'avg_horsepower'
}, inplace=True)
print(average)
```

##9. Exporting Data


```python
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv('high_mpg_cars.csv')
```

##10. Finding Anomalies

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)

IQR = Q3 - Q1
print("IQR of mpg :",IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]
print(outliers)
```

```python
df_outliers = outliers[['car name', 'mpg', 'model year']]

print(df_outliers)
```

#NumPy Problems

```python
import numpy as np
```

##1. Basic Array Operations

```python
mpg_arr=df['mpg'].to_numpy()
print("Mean:",np.mean(mpg_arr))
print("Median:",np.median(mpg_arr))
print("Standard Deviation:",np.std(mpg_arr))
```

```python
car_count=np.sum(mpg_arr > 30)
print("Number of cars with mpg greater than 25 is :", car_count)
```

##2. Filtering

```python
cyl_arr = df['cylinders'].to_numpy()
name_arr = df['car name'].to_numpy()

filtered_names = name_arr[cyl_arr > 6]

result = filtered_names.tolist()
print(result)
```

##3. Statistical Analysis

```python
weight_arr=df['weight'].to_numpy()

per_25=np.percentile(weight_arr,25)
per_50=np.percentile(weight_arr,50)
per_75=np.percentile(weight_arr,75)
print("25th percentile :",per_25)
print("50th percentile :",per_50)
print("75th percentile :",per_75)
```

##4. Array Manipulation

```python
acc_arr=df['acceleration'].to_numpy()

acc_normalized=(acc_arr - acc_arr.min()) / (acc_arr.max() - acc_arr.min())
print(acc_normalized)
```

##5. Broadcasting

```python
hp_arr = df['horsepower'].to_numpy()
updated_hp = hp_arr * 1.10

print(updated_hp)
```

##6. Boolean Indexing

```python
origin_arr = df['origin'].to_numpy()
disp_arr = df['displacement'].to_numpy()

disp_europe = disp_arr[origin_arr == 2]
avg_disp = np.mean(disp_europe)

print("Average displacement (Europe):", avg_disp)
```

##7. Matrix Operations

```python
matrix = df[['mpg', 'horsepower', 'weight']].to_numpy()

vector = np.array([1, 0.5, -0.2])
result = np.dot(matrix, vector)

print(result)
```

##8. Sorting


```python
model_arr = df['model year'].to_numpy()
car_arr = df['car name'].to_numpy()

sorted_years = np.argsort(model_arr)[::-1]

top5_names = car_arr[sorted_years][:5]
print(top5_names)
```

##9. Correlation

```python
mpg_arr = df['mpg'].to_numpy()
weight_arr = df['weight'].to_numpy()

corr_matrix = np.corrcoef(mpg_arr, weight_arr)
corr = corr_matrix[0, 1]

print("Pearson Correlation:", corr)
```

##10.Conditional Aggregates

```python
cyl_array = df['cylinders'].to_numpy()
mpg_array = df['mpg'].to_numpy()

unique_cyl = np.unique(cyl_array)

for cyl in unique_cyl:
    mean_mpg = np.mean(mpg_array[cyl_array == cyl])
    print(f"Cylinders: {cyl}, Mean MPG: {mean_mpg}")
```
