#1. Bank Account Manager

```python
class BankAccount:
    def __init__(self, owner, accno, balance=0):
        self.owner = owner
        self.accno = accno
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("Amount deposited successfully")
        else:
            print("Invalid amount")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print("Withdrawal successful")

    def check_balance(self):
        print("\n--- Account Details ---")
        print("Name:", self.owner)
        print("Account Number:", self.accno)
        print("Balance:", self.balance)


# Create account (no balance input needed)
name = input("Enter name: ")
accno = int(input("Enter account number: "))

account = BankAccount(name, accno)

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Balance Enquiry")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        amt = float(input("Enter amount: "))
        account.deposit(amt)

    elif choice == 2:
        amt = float(input("Enter amount: "))
        account.withdraw(amt)

    elif choice == 3:
        account.check_balance()

    elif choice == 4:
        print("Thank you!")
        break

    else:
        print("Invalid choice")
```

#2.Weather Data Analyzer

```python
class WeatherAnalyzer:
    def __init__(self, region, temperatures, humidity):
        self.region = region
        self.temperatures = temperatures
        self.humidity = humidity

    def calculate_avg_temperature(self):
        return sum(self.temperatures) / len(self.temperatures)

    def calculate_avg_humidity(self):
        return sum(self.humidity) / len(self.humidity)

    def determine_weather(self):
        avg_temp = self.calculate_avg_temperature()
        avg_hum = self.calculate_avg_humidity()

        if avg_temp > 30 and avg_hum < 70:
            return "Sunny"
        elif avg_hum >= 70:
            return "Rainy"
        else:
            return "Cloudy"

    def display_report(self):
        print("\n Weather Report ")
        print("Region:", self.region)
        print("Temperatures:", self.temperatures)
        print("Humidity:", self.humidity)
        print("Average Temperature:", self.calculate_avg_temperature())
        print("Average Humidity:", self.calculate_avg_humidity())
        print("Weather Condition:", self.determine_weather())


#USER INPUT

region = input("Enter region: ")

n = int(input("Enter number of days: "))

temperatures = []
humidity = []

print("\nEnter temperature values:")
for i in range(n):
    temp = float(input(f"Day {i+1}: "))
    temperatures.append(temp)

print("\nEnter humidity values:")
for i in range(n):
    hum = float(input(f"Day {i+1}: "))
    humidity.append(hum)


#Create object
weather = WeatherAnalyzer(region, temperatures, humidity)

#result
weather.display_report()
```
